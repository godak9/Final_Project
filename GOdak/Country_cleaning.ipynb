```python
import pandas as pd 
import numpy as np
import re
```

```python
#Importing sharks data--using file that was cleaned and only contains confirmed shark attacks 
# From kaggle https://www.kaggle.com/datasets/thedevastator/shark-attacks-the-risks-of-coastal-water-activit
# FromShark Research Institute's website https://www.sharkattackfile.net/incidentlog.htm
file_path = "shark_attack_confirmed.csv"
df = pd.read_csv(file_path)
df

#Create df with only country column to work with 
columns = ["Case Number", "Country"]
country_df = df.loc[:, columns]
country_df
```

```python
counts_df = country_df['Country'].value_counts()
pd.set_option('display.max_rows', None)
counts_df
```

```python
#Top five USA, AUSTRALIA, SOUTH AFRICA, PAPUA NEW GUINEA, NEW ZEALAND 
# Add 3 letter country code labels 
country_df.loc[country_df['Country'] == 'USA', 'Country'] = 'USA'
country_df.loc[country_df['Country'] == 'AUSTRALIA', 'Country'] = 'AUS'
country_df.loc[country_df['Country'] == 'SOUTH AFRICA', 'Country'] = 'ZAF'
country_df.loc[country_df['Country'] == 'PAPUA NEW GUINEA', 'Country'] = 'PNG'
country_df.loc[country_df['Country'] == 'NEW ZEALAND', 'Country'] = 'NZL'
country_df.loc[country_df['Country'] == 'BAHAMAS', 'Country'] = 'BHS'
country_df.loc[country_df['Country'] == 'MEXICO', 'Country'] = 'MEX'
country_df.loc[country_df['Country'] == 'FIGI', 'Country'] = 'FGI'
country_df.loc[country_df['Country'] == 'ITALY', 'Country'] = 'ITA'
country_df.loc[country_df['Country'] == 'REUNION', 'Country'] = 'REU'
country_df.loc[country_df['Country'] == 'PHILIPPINES', 'Country'] = 'PHL'
country_df.loc[country_df['Country'] == 'NEW CALEDONIA', 'Country'] = 'NCL'
country_df.loc[country_df['Country'] == 'CUBA', 'Country'] = 'CUB'
country_df.loc[country_df['Country'] == 'EGYPT', 'Country'] = 'EGY'
country_df.loc[country_df['Country'] == 'MOZAMBIQUE', 'Country'] = 'MOZ'
country_df.loc[country_df['Country'] == 'INDIA', 'Country'] = 'IND'
country_df.loc[country_df['Country'] == 'SPAIN', 'Country'] = 'ESP'
country_df.loc[country_df['Country'] == 'PANAMA', 'Country'] = 'PAN'
country_df.loc[country_df['Country'] == 'JAPAN', 'Country'] = 'JPN'
country_df.loc[country_df['Country'] == 'SOLOMON ISLANDS', 'Country'] = 'SLB'
country_df.loc[country_df['Country'] == 'IRAN', 'Country'] = 'IRN'
country_df.loc[country_df['Country'] == 'CROATIA', 'Country'] = 'HRV'
country_df.loc[country_df['Country'] == 'FRENCH POLYNESIA', 'Country'] = 'PYF'
country_df.loc[country_df['Country'] == 'JAMAICA', 'Country'] = 'JAM'
country_df.loc[country_df['Country'] == 'HONG KONG', 'Country'] = 'HKG'
country_df.loc[country_df['Country'] == 'INDONESIA', 'Country'] = 'IDN'
country_df.loc[country_df['Country'] == 'ENGLAND', 'Country'] = 'ENG'

```

```python
country_labels = ['USA', 'AUS', 'ZAF','PNG', 'NZL', 'BHS', 'MEX', 'FGI', 'ITA', 'REU',
                  'PHL','NCL', 'CUB', 'EGY', 'MOZ', 'IND', 'ESP', 'PAN', 'JPN', 'SLB',
                  'IRN', 'HRV','PYF', 'JAM', 'HKG', 'IDN', 'ENG'                 
                 ]
# Change the remaining names countries to "OTH" for other  
for row in country_df.Country:
    if row not in country_labels:
        country_df['Country'].replace(to_replace=row, value='OTH', inplace=True)
list(country_df.Country.unique())
```

```python
country_df.to_csv("location.csv", index=False)
```
