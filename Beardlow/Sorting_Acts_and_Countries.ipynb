```python
import pandas as pd 
import numpy as np
```

```python
file_path = "shark_attack_confirmed.csv"
df = pd.read_csv(file_path)
df
```

```python
df['Country_Count'] = df.groupby('Country')['Country'].transform('count')
df.head()
#unique_countries = pd.DataFrame(df.Country.value_counts())
#unique_countries
#country_counts = df.Country.value_counts().tolist()
#country_counts
#Top five USA, AUSTRALIA, SOUTH AFRICA, PAPUA NEW GUINEA, NEW ZEALAND 
```

```python
df2 = df.loc[(df.Country_Count > 0)]
df2
```

```python
df2.Activity.value_counts()
df2['Activity_Count'] = df.groupby('Activity')['Activity'].transform('count')
df2.Activity.value_counts()
```

```python
df3 = df2.loc[(df2.Activity_Count > 100)]
#df3
```

```python
dumb_df = pd.DataFrame(pd.get_dummies(df3['Fatal (Y/N)']))
#dumb_df
```

```python
df4 = df3.join(dumb_df)
#df4
```

```python
dumb_df2 = pd.DataFrame(pd.get_dummies(df4['Activity']))
dumb_df2
```

```python
# Activities and Fatality encoded at this point
df5 = df4.join(dumb_df2)
df5
```
